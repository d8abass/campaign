### Reading the data

```python
import pandas as pd

df = pd.read_csv('data.csv')
df.dropna()
df
```

### Dropping unncesseary columns

```python
df = df.drop(columns = ['id', 'Unnamed: 32'])
df
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

scaler = StandardScaler()
encoder = LabelEncoder()

scaled_cols = df.columns[1:len(df)]
print(scaled_cols)

df_ref_sc = scaler.fit_transform(df.loc[:, 'radius_mean':'fractal_dimension_worst'])
df_ref_end = encoder.fit_transform(df['diagnosis'])
df.isna().sum()

df_ref_exp = pd.DataFrame(df_ref_sc)
df_ref_exp.columns = scaled_cols
df_ref_exp['diagnosis'] = df_ref_end

print(df_ref_exp)

df_ref_exp.to_csv('data_refined.csv')
```

### Visualization

```python
import seaborn as sns

# use first 4 features only, will take a long time to generate otherwise...

df_ref_exp_ft = df_ref_exp[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']]

sns.pairplot(df_ref_exp_ft)
```

```python
ft_cm = df_ref_exp_ft.corr()

print(ft_cm)

sns.heatmap(ft_cm)
```

```python
sns.boxplot(x = "area_mean", data = df_ref_exp_ft)
```

```python
sns.violinplot(x = 'area_mean', data = df_ref_exp_ft)

ft_cols = df_ref_exp_ft.columns
print(ft_cols)
ft_cols_ls = list(ft_cols)
print(ft_cols_ls)

# area mean especially seems to have outliers, based on the violin plot. 
```
